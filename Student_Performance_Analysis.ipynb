# Student Exam Score Analysis using Python

## Objective
The objective of this project is to analyze students' exam scores using Python.

## Dataset
The dataset contains marks of 10 students in four subjects:

- Physics
- Chemistry
- Maths
- English

## Tools Used
- Python
- Pandas
- Matplotlib
- CSV file
- Jupyter Notebook / VS Code


#### Code/Notebook
- Student_Performance_Analysis.ipynb

```python
# Student Exam Score Analysis Project

import pandas as pd
import matplotlib.pyplot as plt

# a) Read CSV file into pandas DataFrame
df = pd.read_csv("D:\Poonam Docs\Student's_Exam_Scores4.csv")

print("First 10 rows:")
print(df.head())

print("\nColumn names:")
print(df.columns)

print("\nData types:")
print(df.dtypes)


```

```python
# b) Descriptive statistics for each subject
subjects = ["Physics", "Chemistry", "Maths", "English"]

print("\nDescriptive Statistics:")

for subject in subjects:
    print("\nSubject:", subject)
    print("Mean:", df[subject].mean())
    print("Median:", df[subject].median())
    print("Mode:", df[subject].mode()[0])
    print("Minimum:", df[subject].min())
    print("Maximum:", df[subject].max())
    print("Range:", df[subject].max() - df[subject].min())
    print("Variance:", df[subject].var())
    print("Standard Deviation:", df[subject].std())
```

```python
# c) Create Total and Percentage columns
df["Total"] = df["Physics"] + df["Chemistry"] + df["Maths"] + df["English"]
df["Percentage"] = (df["Total"] / 400) * 100

print("\nData with Total and Percentage:")
print(df)

# Find topper and lowest scorer
topper = df[df["Total"] == df["Total"].max()]
lowest_scorer = df[df["Total"] == df["Total"].min()]

print("\nTopper:")
print(topper[["Student_ID", "Student_Name", "Total", "Percentage"]])

print("\nLowest Scorer:")
print(lowest_scorer[["Student_ID", "Student_Name", "Total", "Percentage"]])
```

```python
# Scatter plot of Maths scores
plt.figure(figsize=(7, 5))
plt.scatter(df["Student_ID"], df["Maths"])
plt.xlabel("Student ID")
plt.ylabel("Maths Scores")
plt.title("Scatter Plot of Maths Scores")
plt.show()

# Histogram of Maths scores
plt.figure(figsize=(7, 5))
plt.hist(df["Maths"], bins=5, edgecolor="black")
plt.xlabel("Maths Scores")
plt.ylabel("Frequency")
plt.title("Histogram of Maths Scores")
plt.show()
```

```python
# Compute average marks of each subject
subjects = ["Physics","Chemistry","Maths","English"]
subject_averages = df[subjects].mean()

print("Subject-wise averages:")
print(subject_averages)

# Bar chart
plt.figure(figsize=(7, 5))
subject_averages.plot(kind="bar")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.title("Subject-wise Average Marks")
plt.xticks(rotation=0)
plt.show()
```

```python
# Line chart of Total marks vs Student_ID
plt.figure(figsize=(7, 5))
plt.plot(df["Student_ID"], df["Total"], marker="o")
plt.xlabel("Student ID")
plt.ylabel("Total Marks")
plt.title("Total Marks vs Student ID")
plt.grid(True)
plt.show()
```
